```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
%matplotlib inline
```

```python
data1 = np.loadtxt('result/pthread-rtmlton-0-noise-numonly.txt',unpack=True)
data2 = np.loadtxt('result/pthread-rtmlton-1-noise-numonly.txt',unpack=True)
data3 = np.loadtxt('result/pthread-rtmlton-2-noise-numonly.txt',unpack=True)
data4 = np.loadtxt('result/pthread-rtmlton-3-noise-numonly.txt',unpack=True)
data5 = np.loadtxt('result/pthread-rtmlton-4-noise-numonly.txt',unpack=True)

data6 = np.loadtxt('result/pthread-mlton-0-noise-numonly.txt',unpack=True)
data7 = np.loadtxt('result/pthread-mlton-1-noise-numonly.txt',unpack=True)
data8 = np.loadtxt('result/pthread-mlton-2-noise-numonly.txt',unpack=True)
data9 = np.loadtxt('result/pthread-mlton-3-noise-numonly.txt',unpack=True)
data10 = np.loadtxt('result/pthread-mlton-4-noise-numonly.txt',unpack=True)

def setBoxColors(bp):
    bp['boxes'][0].set(color='#7570b3', linewidth=2)
    plt.setp(bp['caps'][0], color='#7570b3', linewidth=2)
    plt.setp(bp['caps'][1], color='#7570b3', linewidth=2)
    plt.setp(bp['whiskers'][0], color='#7570b3', linewidth=2)
    plt.setp(bp['whiskers'][1], color='#7570b3', linewidth=2)
    plt.setp(bp['medians'][0], color='#7570b3', linewidth=2)

    bp['boxes'][1].set(color='#b2df8a', linewidth=2)
    plt.setp(bp['caps'][2], color='#b2df8a', linewidth=2)
    plt.setp(bp['caps'][3], color='#b2df8a', linewidth=2)
    plt.setp(bp['whiskers'][2], color='#b2df8a', linewidth=2)
    plt.setp(bp['whiskers'][3], color='#b2df8a', linewidth=2)
    plt.setp(bp['medians'][1], color='#b2df8a', linewidth=2)



#RTMLTON
data1 = [x/1000 for x in data1]
data2 = [x/1000 for x in data2]
data3 = [x/1000 for x in data3]
data4 = [x/1000 for x in data4]
data5 = [x/1000 for x in data5]

#MLTON
data6 = [x/1000 for x in data6]
data7 = [x/1000 for x in data7]
data8 = [x/1000 for x in data8]
data9 = [x/1000 for x in data9]
data10 = [x/1000 for x in data10]


medRT = [np.median(np.array(data1)),np.median(np.array(data2)),np.median(np.array(data3)),np.median(np.array(data4)),np.median(np.array(data5))]
medML = [np.median(np.array(data6)),np.median(np.array(data7)),np.median(np.array(data8)),np.median(np.array(data9)),np.median(np.array(data10))]



data0n =[data1,data6]
data1n =[data2,data7]
data2n =[data3,data8]
data3n =[data4,data9]
data4n =[data5,data10]

# Create a figure instance
fig = plt.figure(1, figsize=(12, 9),dpi=120,)

# Create an axes instance
ax = fig.add_subplot(111)

rc = dict(markerfacecolor='#e7298a', marker='o',alpha=0.5)

# Create the boxplot
bp = plt.boxplot(data0n,positions = [1, 2], widths = 0.6, whis='range', showfliers=False)
setBoxColors(bp)



bp = plt.boxplot(data1n,positions = [4, 5], widths = 0.6, whis='range', showfliers=False)
setBoxColors(bp)



bp = plt.boxplot(data2n,positions = [7, 8], widths = 0.6, whis='range', showfliers=False)
setBoxColors(bp)



bp = plt.boxplot(data3n,positions = [10, 11], widths = 0.6, whis='range', showfliers=False)
setBoxColors(bp)


bp = plt.boxplot(data4n,positions = [13, 14], widths = 0.6, whis='range', showfliers=False)

setBoxColors(bp)


    
plt.xlim(0,15)
ax.set_xticklabels(['0','1','2','3','4'])
ax.set_xticks([1.5, 4.5, 7.5,10.5,13.5])


aB, = ax.plot([1,4,7,10,13],medRT,'-',color='red',alpha=0.3)
aC, = ax.plot([2,5,8,11,14],medML,'-',color='red',alpha=0.3)

## Remove top axes and right axes ticks
#ax.get_xaxis().tick_bottom()
#ax.get_yaxis().tick_left()

ax.set_ylabel("Response time ($ms$)",color = '#7570b3',fontsize=10)
ax.set_xlabel("Number of noise threads",color = '#7570b3',fontsize=20)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


# draw temporary red and blue lines and use them to create a legend
hB, = ax.plot([1,1],'-',color='#7570b3')
hR, = ax.plot([1,1],'-',color='#b2df8a')

from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('x-large')

plt.legend((hB, hR),('RTMLton', 'Vanilla MLton'),prop = fontP,loc=2)
hB.set_visible(False)
hR.set_visible(False)


name = "CDsmlgraph"

#plt.savefig(name)

pp = PdfPages("{}.pdf".format(name))
pp.savefig(fig)
pp.close()

plt.show()
```
